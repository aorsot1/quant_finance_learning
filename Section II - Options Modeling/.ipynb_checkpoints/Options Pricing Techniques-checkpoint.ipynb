Option pricing via FFT is characterized by the following set of technique parameters:
* α: damping factor introduced to ensure convergence of inner integral.
* N: number of nodes in algorithm, must be of form: N = 2n
* B: upper bound of pricing integral.
* ∆ν: spacing between nodes in pricing integral.
* ∆k: spacing between nodes on strike spacing grid.

Strengths
* Provides a flexible, efficient method for pricing European options with a wide array of models.
* Prices entire volatility skews with the same effort required to produce single option price.
* Naturally lends itself to a general pricing framework with models easily substituted for each other.

Weaknesses:
* Difficult to extend to non-standard payoffs.
* Impossible to extend to path dependent options.
* Model prices are sensitive to parameters, α, N and B.
* Unable to accurately price highly out-of-the-money options

Variants
* Fractional FFT: A modification of the standard FFT technique where we decouple the strike spacing grid ∆k from the grid related to the integral of the characteristic function ∆ν.
* COS Method: Relies on a Fourier Cosine expansion to using the characteristic function to compute the coefficients in the expansion.
* Saddlepoint Method: Treats option prices as tail probabilities and approximates these via a model’s Cumulant Generating Function.

## Optimization 
Generally speaking optimization problems in finance are characterized by multiple parameters. A few examples of where optimization techniques may
be applied in finance are:
* Calibration: Finding an optimal set of parameters to fit a volatility surface
* Estimation: Finding the best fit line to forecast an asset price
* Asset Allocation: Constructing an optimal portfolio given a set of constraints.
* Yield Curve Construction: Finding the model rates that best match traded market rate products.
* Risk Neutral Density Extraction: Extracting a set of probabilities that best fit market data directly.

We refer to f(x) as the objective function in our optimization. We refer to g(x) ≤ c as the constraints in our optimization
* Lagrange Multipliers
* unconstrained optimization
* Convex Functions

## Calibration of volatility surfaces
There are two general methods for calibrating a volatility surface to a set of market
data:
* Parametric: Begin with a model or stochastic process and find the parameters that lead to the best fit to option prices
* Non-Parametric: Attempt to extract a set of probabilities directly

Optimization algorithms can be classified into two main groups: gradient-based and
gradient-free.
* Gradient-based methods utilize information about the function and its derivative in order to determine each iterative step.
* Gradient-free methods utilize only information about the function in order to determine the next step.
